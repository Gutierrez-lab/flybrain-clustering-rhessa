# Querying Mushroom body types 
The types were taken from Scheffer et al. and more. 

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from neuprint import Client, fetch_neurons, NeuronCriteria as NC, fetch_adjacencies


auth_token_file = open("/Users/rweberla/flybrain-clustering-rhessa/flybrain.auth.txt", 'r')
auth_token = next(auth_token_file).strip()
try:
    np_client = Client('neuprint.janelia.org', dataset='hemibrain:v1.2.1', token=auth_token)
except:
    np_client = None
```

```python
# Specifying the neuron criteria for the mushroom body neurons and traced/non-cropped neurons
mb_criteria = NC(status='Traced',type=['MB.*', 'APL', 'KC.*', 'DP.*', 'PAM.*', 'PPL.*', 'OA-VPM3',
                                       'OA-VPM4', 'SIFa', 'ORN_D', 'DA.*', 'DC1.*', 'DC2.*','DC3.*',
                                       'DC4.*', 'DM.*', 'V_.*','VA.*','VC1.*', 'VC2.*','VC3.*',
                                       'VC4.*', 'VC5.*', 'VL.*', 'VM.*'], cropped=False)
mb_neurons, roi_count_df  = fetch_neurons(mb_criteria)
```
