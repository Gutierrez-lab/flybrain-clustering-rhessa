```python
# Importing upstream and downstream neurons
crit = NC(bodyId = 425790257)
n_df_in, conn_df_in = fetch_adjacencies(crit, None)
n_df_o, conn_df_o = fetch_adjacencies(None, crit)

# Sorting neuron lists by weight
conn_df_in = conn_df_in.sort_values(by='weight', ascending= False)
conn_df_o = conn_df_o.sort_values(by='weight', ascending= False)
```

```python
# Seperating into lists and appending from upstream and downstream dataframes
weights = conn_df_o.weight._append(conn_df_in.weight)
neuron_ids = conn_df_o.bodyId_pre._append(conn_df_in.bodyId_post)
apl = conn_df_o.bodyId_post._append(conn_df_in.bodyId_pre)
```

```python
# Creating singular dataframe to pull undirected list from
dict = {"NeuronIds": neuron_ids, 'apl': apl, 'Weights': weights}
df = pandas.DataFrame(dict)
```
