## Creating process for program with module 1

```python
oviIN_df
oviIN_df = oviIN_df[oviIN_df['0.0']==1]
oviIN_df = oviIN_df[['0.0','0.05','0.1','0.5','0.75', '1.0']]
oviIN_df
```

```python
# append _r0.0 to values in column '0.0' and so on
oviIN_df['0.0'] = oviIN_df['0.0'].astype(str) + '_r0.0'
oviIN_df['0.05'] = oviIN_df['0.05'].astype(str) + '_r0.05'
oviIN_df['0.1'] = oviIN_df['0.1'].astype(str) + '_r0.1'
oviIN_df['0.5'] = oviIN_df['0.5'].astype(str) + '_r0.5'
oviIN_df['0.75'] = oviIN_df['0.75'].astype(str) + '_r0.75'
oviIN_df['1.0'] = oviIN_df['1.0'].astype(str) + '_r1.0'
oviIN_df['counts'] =  1
oviIN_df
```

```python
# Sankey organization for 0.0 to 0.05
cluster1_1 = oviIN_df[['0.0','0.05','counts']].groupby(['0.0', '0.05']).count().reset_index()
cluster1_1.columns = ['source','target','value']
cluster1_1
```

```python
# Sankey organization for resolution to resolution+1
cluster1_2 = oviIN_df[['0.05','0.1','counts']].groupby(['0.05', '0.1']).count().reset_index()
cluster1_2.columns = ['source','target','value']

cluster1_3 = oviIN_df[['0.1','0.5','counts']].groupby(['0.1','0.5']).count().reset_index()
cluster1_3.columns = ['source','target','value']

cluster1_4 = oviIN_df[['0.5','0.75','counts']].groupby(['0.5','0.75']).count().reset_index()
cluster1_4.columns = ['source','target','value']

cluster1_5 = oviIN_df[['0.75','1.0','counts']].groupby(['0.75','1.0']).count().reset_index()
cluster1_5.columns = ['source','target','value']

links = pd.concat([cluster1_1, cluster1_2, cluster1_3, cluster1_4, cluster1_5], axis=0)
```

```python
# find the nodes to map onto
nodes = oviIN_df['0.0'].unique().tolist() + oviIN_df['0.05'].unique().tolist() + oviIN_df['0.1'].unique().tolist() + oviIN_df['0.5'].unique().tolist() + oviIN_df['0.75'].unique().tolist() + oviIN_df['1.0'].unique().tolist()
mapping_dict = {k: v for v, k in enumerate(nodes)}

# replace source and target with enumerated values
links['source'] = links['source'].map(mapping_dict)
links['target'] = links['target'].map(mapping_dict)

# to list for sankey diagram
links_dict = links.to_dict(orient='list')
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness=20,
        line=dict(color='blue', width=0.5),
        label = nodes,
        color='green'
    ),
    link = dict(
        source= links_dict['source'],
        target = links_dict['target'],
        value = links_dict['value']
    )
    
)
])

fig.update_layout(height = 500 ,title='Module 1 of oviINr network in the full ovi modularity data across resolutions')
fig.show()
```

```python
cluster_list = [1,2,3,4,5,6]
```

```python
# Create main function that plots all at once
def main(oviIN_df):
    for i in cluster_list:
        oviIN_df_c = oviIN_df[oviIN_df['0.0']==i]
        oviIN_df_c = oviIN_df_c[['0.0','0.05','0.1','0.5','0.75', '1.0']]
        
        oviIN_df_c['0.0'] = oviIN_df_c['0.0'].astype(str) + '_r0.0'
        oviIN_df_c['0.05'] = oviIN_df_c['0.05'].astype(str) + '_r0.05'
        oviIN_df_c['0.1'] = oviIN_df_c['0.1'].astype(str) + '_r0.1'
        oviIN_df_c['0.5'] = oviIN_df_c['0.5'].astype(str) + '_r0.5'
        oviIN_df_c['0.75'] = oviIN_df_c['0.75'].astype(str) + '_r0.75'
        oviIN_df_c['1.0'] = oviIN_df_c['1.0'].astype(str) + '_r1.0'
        oviIN_df_c['counts'] =  1

        log_msg('Creating grouped df for cluster {}'.format(i))
        cluster1_1 = oviIN_df_c[['0.0','0.05','counts']].groupby(['0.0', '0.05']).count().reset_index()
        cluster1_1.columns = ['source','target','value']
        
        cluster1_2 = oviIN_df_c[['0.05','0.1','counts']].groupby(['0.05', '0.1']).count().reset_index()
        cluster1_2.columns = ['source','target','value']

        cluster1_3 = oviIN_df_c[['0.1','0.5','counts']].groupby(['0.1','0.5']).count().reset_index()
        cluster1_3.columns = ['source','target','value']

        cluster1_4 = oviIN_df_c[['0.5','0.75','counts']].groupby(['0.5','0.75']).count().reset_index()
        cluster1_4.columns = ['source','target','value']

        cluster1_5 = oviIN_df_c[['0.75','1.0','counts']].groupby(['0.75','1.0']).count().reset_index()
        cluster1_5.columns = ['source','target','value']

        links = pd.concat([cluster1_1, cluster1_2, cluster1_3, cluster1_4, cluster1_5], axis=0)

        # find the nodes to map onto
        nodes = oviIN_df_c['0.0'].unique().tolist() + oviIN_df_c['0.05'].unique().tolist() + oviIN_df_c['0.1'].unique().tolist() + oviIN_df_c['0.5'].unique().tolist() + oviIN_df_c['0.75'].unique().tolist() + oviIN_df_c['1.0'].unique().tolist()
        mapping_dict = {k: v for v, k in enumerate(nodes)}

        # replace source and target with enumerated values
        links['source'] = links['source'].map(mapping_dict)
        links['target'] = links['target'].map(mapping_dict)

        # to list for sankey diagram
        links_dict = links.to_dict(orient='list')

        fig = go.Figure(data=[go.Sankey(
            node = dict(
                pad = 15,
                thickness=20,
                line=dict(color='blue', width=0.5),
                label = nodes,
                color='green'
            ),
            link = dict(
                source= links_dict['source'],
                target = links_dict['target'],
                value = links_dict['value']
            )
            
        )
        ])

        fig.update_layout(height = 800 ,title='Module {} of oviINr network in the full ovi modularity data across resolutions'.format(i))
        fig.show()        
```

```python
main(oviIN_df)
```
