```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
plt.rcParams["pdf.fonttype"] = 42
plt.rcParams["font.sans-serif"] = "Helvetica"
```

```python
tick_kws = dict(fontsize=8)
label_kws = dict(fontsize=10)
title_kws = dict(fontsize=12)


plt.figure(figsize=(3.35,5.025), tight_layout="tight")

robustness = {}
for plot, chi in enumerate(["000","025"]):
    with open(f"analysis/v1.1/robustness-chi-{chi}", "r") as file:
        robustness[chi] = np.array([[i, int(num.strip())] for i, num in enumerate(file.readlines())])
    plt.subplot(2,1,plot + 1)
    plt.plot(robustness[chi][:,0], robustness[chi][:,1],
             marker=".", color="black",
             linewidth=0.5)
    plt.yscale("log")
    plt.title(r"$\chi$ = " + chi[0] + "." + chi[1:], **title_kws)
    plt.xticks(**tick_kws)
    plt.yticks(**tick_kws)
    plt.minorticks_off()
    plt.ylabel("Number of Pairs", labelpad=-1, **label_kws)
    if plot == 1:
        plt.xlabel("Frequency of Co-Clustering", **label_kws)
        
for ext in ["png","pdf"]:
    plt.savefig(os.path.join(figure_dir, "Supplement", f"CBF-vs-RenEEL.{ext}"), bbox_inches="tight")
```
